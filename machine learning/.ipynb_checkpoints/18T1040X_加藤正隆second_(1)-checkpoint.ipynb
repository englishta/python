```python
import pyper
r = pyper.R()
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
r.assign("x", x)
r("ans <- sum(x)")
```

```python
ans2 = r.get("ans1")
```

```python
import pyper
r = pyper.R()
x = [0]*9
for i in range(9):
    x[i]=i+1
print("x = ", x)
t = sum(x)
print("sum(x) = ", t)
r("ans1 <- t")
ans2 = r.get("ans1")
ans2
```
